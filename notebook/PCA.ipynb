```python
%matplotlib inline
from sklearn.decomposition import PCA

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
with open('../data/posture_train.txt', 'r') as rf:
    data = pd.read_csv(rf, delim_whitespace=True, header=None)
data.head()
```

```python
x = []
y = []
for index, row in data.iterrows():
    x.append(np.array(row[:4]))
    for action, value in enumerate(row[4:]):
        if value == 1:
            y.append(action)
            break;
    
x_data = np.array(x)
y_data = np.array(y)
```

```python
pca = PCA(n_components=2)
#processed = pca.fit_transform(x_data)
pca.fit(x_data)
result = pca.transform(x_data)
print(result)
```

```python
print(pca.components_)
```

```python
plt.style.use('ggplot')
plt.scatter(result[:,0], result[:,1], c=y_data, s=25, alpha=0.4, marker='o')
```
